```python
# write a "?" (after the optional letter) to keep one specific letter of a string as optional
```

```python
customer_reviews = ['sam was a great help to me in the store', 
 'the cashier was very rude to me, I think her name was eleanor', 
 'amazing work from sadeen!', 
 'sarah was able to help me find the items i needed quickly', 
 'lucy is such a great addition to the team', 
 'great service from sara she found me what i wanted']
```

```python
sarahs_reviews = []
```

```python
pattern_to_find = r"sarah?"
```

```python
for string in customer_reviews:
    if (re.search(pattern_to_find,string)):
        sarahs_reviews.append(string)
        
```

```python
print(sarahs_reviews)
```

```python
a_reviews = []
pattern_to_find = r"^a"
```

```python
for string in customer_reviews:
    if(re.search(pattern_to_find,string)):
        a_reviews.append(string)
```

```python
print(a_reviews)
```

```python
#"^Char" - start of the string and "char$" - end of the string
```

```python
#| - works like an or 
```

```python
needwant_reviews = []
```

```python
pattern_to_find = r"(need|want)ed"
```

```python
for string in customer_reviews:
    if(re.search(pattern_to_find,string)):
        needwant_reviews.append(string)
```

```python
print(needwant_reviews)
```

```python
#STEMMING - reduces complexity by reducing unique words from the dataset by standardizing the text
```

```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
connect_tokens = ['connecting']
```

```python
connect_tokens = ['connecting', 'connected',  'connectivity', 'connect', 'connects']
```

```python
for t in connect_tokens:
    print(t, ":", ps.stem(t))
```

```python
#LEMMATIZATION
```

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
for t in connect_tokens:
    print(t,": ", lemmatizer.lemmatize(t))
```
