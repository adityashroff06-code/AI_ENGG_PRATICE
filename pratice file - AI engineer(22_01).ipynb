```python
def add_10(m):          #Functions with conditional statements within itself
    if m>=100:
        m = m+10
        return m
    else:
        return "save more"
```

```python
def count(numbers):
    total = 0
    for x in numbers:
        if x<20:
            total = total + 1
    return total 
```
